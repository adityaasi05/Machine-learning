```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
df = pd.read_csv('placement_data.csv')
```

```python
df.head()
```

```python
df.iloc[:,1]
```

```python
plt.scatter(df['cgpa'],df['package'])
plt.xlabel('CGPA')
plt.ylabel('Package(in lpa)')
plt.show()
```

```python
df.iloc[:,0:1]# CGPA
```

```python
X = df.iloc[:,0:1]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
lr.predict(X_test.iloc[0].values.reshape(1,1))
```

```python
plt.scatter(df['cgpa'],df['package'])
plt.plot(X_train,lr.predict(X_train),color='red')
plt.xlabel('CGPA')
plt.ylabel('Package(in lpa)')
```

```python
m = lr.coef_
```

```python
b = lr.intercept_
```

```python
print("Coef :",m)
print("Intecept",b)
```

```python
data = {
    'Size_sqft': [600, 800, 1000, 1200, 1400, 1600, 1800, 2000, 2200, 2400],
    'Price_1000s': [150, 180, 210, 240, 270, 300, 330, 360, 390, 420]
}

df = pd.DataFrame(data)
df
```

```python
plt.scatter(df['Size_sqft'],df['Price_1000s'])
plt.xlabel('CGPA')
plt.ylabel('Package(in lpa)')
```

```python
x=df['Size_sqft']
y=df['Price_1000s']   
```

```python
lr = LinearRegression()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
lr.fit(x_train.values.reshape(-1,1),y_train)
```

```python
lr.predict(x_test.iloc[0].reshape(1,1))
```

```python
plt.scatter(df['Size_sqft'],df['Price_1000s'])
plt.plot(x_train,lr.predict(x_train.values.reshape(-1,1)),color='red')
plt.xlabel('Size_sqft')
plt.ylabel('Price_1000s')
```
